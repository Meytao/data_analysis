# Pandas Series 入门教程

Pandas 是 Python 数据分析领域最核心的库之一，它提供了高效的数据结构和数据分析工具。在正式开始学习 Series 之前，我们先来了解一下 Pandas 在数据处理生态系统中的位置。

如果你之前使用过 Excel，可能会习惯于在电子表格中处理数据。Pandas 可以看作是一个更加强大的电子表格，但它的设计初衷是处理大规模数据，并且可以与 Python 的其他数据处理工具完美配合。类似于 SQL 中的 SELECT、WHERE、GROUP BY 等操作，在 Pandas 中都可以通过简洁的代码实现。

**Series 是 Pandas 中最基本的数据结构，它可以看作是一维数组**，类似于 Excel 中的一列数据。与普通数组不同的是，Series 带有索引（Index），这使得数据访问更加灵活和直观。接下来，我们将系统地学习 Series 的各种操作。

在开始之前，请确保已经导入了必要的库：

```python
import pandas as pd
import numpy as np
```

## 1. Series 的创建

创建 Series 有多种方式，我们可以从列表、字典创建，也可以指定索引。理解不同的创建方式有助于我们根据实际数据选择最合适的方法。

### 1.1 从列表创建

最简单的创建方式是直接从 Python 列表创建 Series。此时，Pandas 会自动为每个元素分配一个整数索引（从 0 开始）。

```python
# 从列表创建 Series
s = pd.Series([10, 20, 30, 40, 50])
print("基本 Series:")
print(s)
print("\n索引:", s.index.tolist())
print("值:", s.values)
```

### 1.2 自定义索引

我们可以为 Series 指定自定义索引，这使得数据访问更加语义化。自定义索引可以是字符串、数字或其他可哈希的类型。

```python
# 使用自定义索引
s = pd.Series([10, 20, 30, 40, 50], 
               index=["A", "B", "C", "D", "E"])
print("带自定义索引的 Series:")
print(s)
print("\n索引:", s.index.tolist())
```

### 1.3 从字典创建

字典是一种键值对的数据结构，可以直接转换为 Series。此时，字典的键会成为 Series 的索引，字典的值会成为 Series 的值。

```python
# 从字典创建 Series
data = {"苹果": 3, "香蕉": 2, "橙子": 5, "葡萄": 4}
s = pd.Series(data)
print("从字典创建的 Series:")
print(s)
```

### 1.4 指定 name 属性

Series 有一个 name 属性，用于描述这个 Series 所代表的数据类型。这在数据分析和数据交换时非常有用，可以帮助我们理解数据的含义。

```python
# 创建带 name 的 Series
s = pd.Series([10, 20, 30, 40, 50], 
               index=["A", "B", "C", "D", "E"], 
               name="月份销量")
print("带名称的 Series:")
print(s)
print("\nSeries 名称:", s.name)
```

### 1.5 带重复索引的创建

Series 允许索引重复，这在某些场景下非常有用，比如表示多组测量值与同一个标签的对应关系。

```python
# 带重复索引的 Series
s = pd.Series([100, 200, 300], index=["一月", "一月", "二月"])
print("带重复索引的 Series:")
print(s)
```

## 2. Series 的属性

Series 提供了多个属性来访问其基本信息。了解这些属性有助于我们更好地理解数据结构。

- **index**: 返回 Series 的索引对象
- **values**: 返回 Series 的值（作为 numpy 数组）
- **dtype**: 返回 Series 元素的数据类型
- **shape**: 返回 Series 的形状（一个元组）
- **ndim**: 返回 Series 的维度（始终为 1）
- **size**: 返回 Series 的元素个数
- **name**: 返回 Series 的名称

```python
# 创建示例 Series
s = pd.Series([10, 20, 30, 40, 50], 
               index=["A", "B", "C", "D", "E"],
               name="示例数据")

print("=== Series 属性 ===")
print(f"索引 (index): {s.index.tolist()}")
print(f"值 (values): {s.values}")
print(f"数据类型 (dtype): {s.dtype}")
print(f"形状 (shape): {s.shape}")
print(f"维度 (ndim): {s.ndim}")
print(f"元素个数 (size): {s.size}")
print(f"名称 (name): {s.name}")
```

## 3. Series 的索引

Series 提供了多种索引方式，可以分为显式索引和隐式索引两大类。理解这两种索引方式的区别对于熟练使用 Series 至关重要。

### 3.1 显式索引（loc）

显式索引使用我们自定义的标签来访问数据。当我们创建 Series 时指定的索引（如 A, B, C 或日期等），就是显式索引。通过 loc 属性可以按显式索引进行数据访问。

```python
s = pd.Series([10, 20, 30, 40, 50], 
               index=["A", "B", "C", "D", "E"])

# 使用 loc 进行显式索引访问
print("使用 loc[\"C\"]:", s.loc["C"])       # 单个元素
print("使用 loc[[\"A\", \"C\"]]:", s.loc[["A", "C"]])  # 多个元素
print("使用 loc[\"A\":\"C\"]:", s.loc["A":"C"])      # 切片（包含两端）
```

### 3.2 隐式索引（iloc）

隐式索引是 Pandas 自动为元素分配的整数索引，从 0 开始。通过 iloc 属性可以按隐式索引进行数据访问，这与 Python 列表的索引方式类似。

```python
# 使用 iloc 进行隐式索引访问
print("使用 iloc[0]:", s.iloc[0])        # 单个元素
print("使用 iloc[[0, 2]]:", s.iloc[[0, 2]])  # 多个元素
print("使用 iloc[0:3]:", s.iloc[0:3])     # 切片（不包含结束位置）
```

### 3.3 单元素快速访问（at / iat）

当需要访问单个元素时，at 和 iat 比 loc 和 iloc 更快。at 用于显式索引，iat 用于隐式索引。

```python
# 使用 at 和 iat 快速访问单个元素
print("使用 at[\"B\"]:", s.at["B"])   # 显式索引
print("使用 iat[1]:", s.iat[1])    # 隐式索引
```

### 3.4 布尔索引

布尔索引允许我们根据条件筛选数据。通过创建一个布尔 Series（条件表达式的结果），可以过滤出满足条件的数据元素。

```python
# 布尔索引示例
s = pd.Series([10, 25, 30, 45, 50, 65, 70])

# 筛选大于 30 的元素
print("大于 30 的元素:")
print(s[s > 30])

# 筛选在 20 到 50 之间的元素
print("\n在 20 到 50 之间的元素:")
print(s[(s >= 20) & (s <= 50)])
```

### 3.5 直接索引

除了使用 loc 和 iloc，还可以直接使用中括号进行索引。这种方式会先尝试显式索引，如果找不到则回退到隐式索引。建议在明确场景下优先使用 loc 和 iloc 以避免歧义。

```python
s = pd.Series([10, 20, 30, 40, 50], 
               index=[1, 2, 3, 4, 5])

# 直接使用中括号索引
print("s[1]:", s[1])        # 隐式索引
print("s[1:3]:", s[1:3])    # 切片
```

## 4. Series 的运算

Series 支持丰富的运算操作，包括标量运算、元素级运算和索引对齐运算。理解这些运算方式对于数据处理非常重要。

### 4.1 标量运算

标量运算指的是 Series 与单个数值之间的运算。运算会应用到 Series 的每个元素上，这是一种向量化操作，无需编写循环。

```python
s = pd.Series([10, 20, 30, 40, 50])

print("原 Series:")
print(s)
print("\n加法 s + 5:")
print(s + 5)
print("\n乘法 s * 2:")
print(s * 2)
print("\n幂运算 s ** 2:")
print(s ** 2)
```

### 4.2 元素级运算

元素级运算允许我们对两个 Series 的对应元素进行运算。当两个 Series 长度不同时，缺失的位置会填充 NaN 值。

```python
s1 = pd.Series([10, 20, 30])
s2 = pd.Series([1, 2, 3])

print("s1:", s1.values)
print("s2:", s2.values)
print("\ns1 + s2:", (s1 + s2).values)
print("s1 * s2:", (s1 * s2).values)
```

### 4.3 索引对齐

当两个 Series 进行运算时，Pandas 会根据索引自动对齐数据。如果某个索引只存在于一个 Series 中，结果中对应的位置会是 NaN。

```python
s1 = pd.Series([10, 20, 30], index=["A", "B", "C"])
s2 = pd.Series([1, 2, 3], index=["B", "C", "D"])

print("s1:", s1.to_dict())
print("s2:", s2.to_dict())
print("\ns1 + s2 (索引对齐后):")
print(s1 + s2)
```

## 5. Series 的常用方法

Series 提供了大量内置方法用于数据处理和分析。下面我们详细介绍各种常用方法的使用。

### 5.1 描述性统计方法

描述性统计方法用于快速了解数据的基本特征，包括集中趋势、离散程度等。

```python
# 创建示例数据
s = pd.Series([10, 20, 30, 40, 50])

print("=== 描述性统计 ===")
print(f"sum()   - 求和: {s.sum()}")
print(f"mean()  - 平均值: {s.mean()}")
print(f"median()- 中位数: {s.median()}")
print(f"std()   - 标准差: {s.std()}")
print(f"var()   - 方差: {s.var()}")
print(f"min()   - 最小值: {s.min()}")
print(f"max()   - 最大值: {s.max()}")
```

```python
# describe 方法一次性获取所有描述性统计信息
s = pd.Series([10, 20, 30, 40, 50])
print("describe() 输出:")
print(s.describe())
```

```python
# 创建包含重复值的 Series 用于演示其他统计方法
s = pd.Series([1, 2, 2, 3, 3, 3, 4, 4, 4, 4])

print("=== 其他统计方法 ===")
print(f"value_counts() - 计数:\n{s.value_counts()}")
print(f"\nquantile(0.25) - 25%分位数: {s.quantile(0.25)}")
print(f"quantile(0.5)  - 50%分位数: {s.quantile(0.5)}")
print(f"quantile(0.75) - 75%分位数: {s.quantile(0.75)}")
print(f"\nmode() - 众数: {s.mode().tolist()}")
```

### 5.2 缺失值处理

现实数据中经常存在缺失值，Pandas 提供了多种检测和处理缺失值的方法。

在 Pandas 中，缺失值用 NaN（Not a Number）表示，这是浮点数类型中的一个特殊值。对于整数类型的数据，缺失值也会被存储为浮点数形式。

```python
# 创建包含缺失值的 Series
s = pd.Series([10, np.nan, 30, None, 50])
print("包含缺失值的 Series:")
print(s)

print("\nisna() - 检测每个元素是否为缺失值:")
print(s.isna())

print("\nnotna() - 检测每个元素是否不是缺失值:")
print(s.notna())
```

```python
# 处理缺失值的方法
s = pd.Series([10, np.nan, 30, None, 50])

print("原 Series:")
print(s)

print("\ndropna() - 删除缺失值:")
print(s.dropna())

print("\nfillna(0) - 用 0 填充缺失值:")
print(s.fillna(0))

print("\nffill() - 用前向值填充:")
print(s.ffill())

print("\nbfill() - 用后向值填充:")
print(s.bfill())
```

### 5.3 排序方法

排序是数据分析中非常常见的操作。Series 提供了按值排序和按索引排序两种方式。

```python
s = pd.Series([30, 10, 50, 20, 40], index=["C", "A", "E", "B", "D"])
print("原 Series:")
print(s)

print("\nsort_values() - 按值升序排序:")
print(s.sort_values())

print("\nsort_values(ascending=False) - 按值降序排序:")
print(s.sort_values(ascending=False))

print("\nsort_index() - 按索引排序:")
print(s.sort_index())
```

### 5.4 唯一值和去重

处理数据时，经常需要找出唯一值或去除重复值。Series 提供了相应的方法来支持这些操作。

```python
s = pd.Series([1, 2, 2, 3, 3, 3, 4, 4, 4, 4])

print("原 Series:")
print(s)

print("\nunique() - 获取唯一值:")
print(s.unique())

print(f"\nnunique() - 唯一值个数: {s.nunique()}")

print("\nvalue_counts() - 计数（按值降序）:")
print(s.value_counts())
```

### 5.5 替换操作

replace 方法可以替换 Series 中的值。它支持单个值替换、多个值替换以及映射替换。

```python
s = pd.Series([10, 20, 30, 40, 50])

print("原 Series:")
print(s)

print("\n替换单个值 (20 -> 200):")
print(s.replace(20, 200))

print("\n替换多个值 ({20: 200, 30: 300}):")
print(s.replace({20: 200, 30: 300}))
```

### 5.6 成员资格判断

isin 方法用于检查 Series 中的每个元素是否属于指定的值集合。这类似于 SQL 中的 IN 语句。

```python
s = pd.Series([10, 20, 30, 40, 50])

print("原 Series:")
print(s)

print("\nisin([20, 40, 60]) - 检查元素是否在指定集合中:")
print(s.isin([20, 40, 60]))

print("\n筛选出在集合中的元素:")
print(s[s.isin([20, 40, 60])])
```

### 5.7 头部和尾部

head 和 tail 方法分别用于获取 Series 的前 n 行和后 n 行，这在查看大数据集时非常有用。

```python
s = pd.Series(range(1, 101))

print("head(5) - 前 5 个元素:")
print(s.head(5))

print("\ntail(5) - 后 5 个元素:")
print(s.tail(5))
```

### 5.8 应用函数

apply 方法允许我们对 Series 的每个元素应用一个函数。这是一种灵活的数据转换方式。

```python
s = pd.Series([1, 2, 3, 4, 5])

print("原 Series:")
print(s)

print("\napply(lambda x: x ** 2) - 每个元素平方:")
print(s.apply(lambda x: x ** 2))

# 更复杂的例子：将数值转换为分类
print("\n应用自定义函数（成绩分级）:")
def grade(score):
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    else:
        return "D"

scores = pd.Series([85, 92, 78, 65, 88])
print("成绩:", scores.tolist())
print("等级:", scores.apply(grade).tolist())
```

### 5.9 累计计算

累计计算用于计算数据的累计值，在分析趋势时非常有用。

```python
s = pd.Series([10, 20, 30, 40, 50])

print("原 Series:")
print(s)

print("\ncumsum() - 累计和:")
print(s.cumsum())

print("\ncumprod() - 累计积:")
print(s.cumprod())

print("\ncummin() - 累计最小值:")
print(s.cummin())

print("\ncummax() - 累计最大值:")
print(s.cummax())
```

### 5.10 变化率计算

pct_change 方法计算相邻元素之间的百分比变化，常用于分析时间序列数据的增长率。

```python
prices = pd.Series([100, 105, 102, 110, 115])

print("价格 Series:")
print(prices)

print("\npct_change() - 价格变化率:")
print(prices.pct_change())

print("\n解释: 第二天的变化率 = (105-100)/100 = 0.05 = 5%")
```

## 6. 实战示例

下面通过几个实际案例来综合运用 Series 的各项操作。

### 6.1 学生成绩分析

创建一个包含 10 名学生数学成绩的 Series，分析成绩分布情况。

```python
import pandas as pd
import numpy as np

# 设置随机种子以保证结果可复现
np.random.seed(42)

# 生成 10 名学生的成绩
values = np.random.randint(50, 101, 10)
indexes = [f"学生{i}" for i in range(1, 11)]
scores = pd.Series(values, indexes, name="数学成绩")

print("=== 学生成绩分析 ===")
print("\n成绩单:")
print(scores)

print(f"\n平均分: {scores.mean():.2f}")
print(f"最高分: {scores.max()}")
print(f"最低分: {scores.min()}")
print(f"中位数: {scores.median()}")
print(f"标准差: {scores.std():.2f}")

# 找出高于平均分的学生
mean_score = scores.mean()
above_avg = scores[scores > mean_score]
print(f"\n高于平均分({mean_score:.2f})的学生:")
print(above_avg)
print(f"高于平均分的学生人数: {len(above_avg)}")
```

### 6.2 温度数据统计

分析某城市一周每天的最高温度。

```python
import pandas as pd
import numpy as np

# 创建温度数据
temperatures = pd.Series(
    [28, 31, 29, 32, 30, 27, 33],
    index=["周一", "周二", "周三", "周四", "周五", "周六", "周日"]
)

print("=== 温度数据统计 ===")
print("\n一周温度:")
print(temperatures)

# 找出温度超过 30 度的天数
hot_days = temperatures[temperatures > 30]
print(f"\n超过 30 度的天数: {len(hot_days)}")
print("这些天是:", hot_days.index.tolist())

# 计算平均温度
print(f"\n平均温度: {temperatures.mean():.2f}°C")

# 将温度从高到低排序
sorted_temps = temperatures.sort_values(ascending=False)
print("\n温度从高到低排序:")
print(sorted_temps)

# 找出温度变化最大的两天
temp_diff = temperatures.diff().abs()
max_change_idx = temp_diff.idxmax()
print(f"\n温度变化最大的两天: {max_change_idx}")
```

### 6.3 股票价格分析

分析股票每日收益率和波动率。

```python
import pandas as pd
import numpy as np

# 创建股票价格数据
prices = pd.Series(
    [102.3, 103.5, 105.1, 104.8, 106.2, 107.0, 106.5, 108.1, 109.3, 110.2],
    index=pd.date_range("2023-01-01", periods=10)
)

print("=== 股票价格分析 ===")
print("\n每日收盘价:")
print(prices)

# 计算每日收益率
returns = prices.pct_change()
print("\n每日收益率:")
print(returns)

# 找出收益率最高和最低的日期
print(f"\n收益率最高的日期: {returns.idxmax().date()}")
print(f"收益率最低的日期: {returns.idxmin().date()}")

# 计算波动率（收益率的标准差）
volatility = returns.std()
print(f"\n波动率（收益率标准差）: {volatility:.6f}")
```

### 6.4 销售数据分析

分析产品月度销售数据，计算季度平均和环比增长率。

```python
import pandas as pd
import numpy as np

# 创建月度销售数据
sales = pd.Series(
    [120, 135, 145, 160, 155, 170, 180, 175, 190, 200, 210, 220],
    index=pd.date_range("2022-01-01", periods=12, freq="MS")
)

print("=== 销售数据分析 ===")
print("\n月度销售数据:")
print(sales)

# 计算季度平均销量（按季度重采样）
quarterly_avg = sales.resample("QS").mean()
print("\n季度平均销量:")
print(quarterly_avg)

# 找出销量最高的月份
print(f"\n销量最高的月份: {sales.idxmax().strftime("%Y年%m月")}")
print(f"最高销量: {sales.max()}")

# 计算月环比增长率
growth_rate = sales.pct_change()
print("\n月环比增长率:")
for i, (date, rate) in enumerate(growth_rate.items()):
    if pd.notna(rate):
        print(f"{date.strftime("%Y年%m月")}: {rate*100:.2f}%")
```

### 6.5 每小时销售额分析

分析商店每小时的销售额数据。

```python
import pandas as pd
import numpy as np

# 设置随机种子
np.random.seed(42)

# 创建每小时销售额数据
hourly_sales = pd.Series(
    np.random.randint(0, 100, 24),
    index=pd.date_range("2025-01-01", periods=24, freq="h")
)

print("=== 每小时销售额分析 ===")
print("\n每小时销售额:")
print(hourly_sales)

# 按天重采样计算每日总销售额
daily_sales = hourly_sales.resample("D").sum()
print("\n每日总销售额:")
print(daily_sales)

# 计算营业时间和非营业时间的销售额
# 营业时间: 8:00-22:00
business_hours = hourly_sales[(hourly_sales.index.hour >= 8) & (hourly_sales.index.hour <= 22)]
non_business_hours = hourly_sales[(hourly_sales.index.hour < 8) | (hourly_sales.index.hour > 22)]

print(f"\n营业时间总销售额: {business_hours.sum()}")
print(f"非营业时间总销售额: {non_business_hours.sum()}")
print(f"营业时间占比: {business_hours.sum() / hourly_sales.sum() * 100:.2f}%")

# 找出销售额最高的 3 个小时
top_3_hours = hourly_sales.nlargest(3)
print("\n销售额最高的 3 个小时:")
for time, value in top_3_hours.items():
    print(f"  {time.strftime("%H:00")}: {value}")
```

- Series 是一种带索引的一维数组，可以通过列表、字典或标量值创建
- 索引分为显式索引（使用 loc）和隐式索引（使用 iloc）
- 布尔索引是数据筛选的强大工具
- 运算时会自动按索引对齐数据
- 丰富的内置方法支持描述性统计、缺失值处理、排序、去重等操作

建议多加练习，将这些概念应用到实际数据分析中，逐步熟练掌握 Series 的使用。
